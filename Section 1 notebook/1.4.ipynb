```python
import pandas as pd
from sklearn import preprocessing
from sklearn.preprocessing import Binarizer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('WA_Fn-UseC_-Sales-Win-Loss.csv')
```

```python
data.describe()
```

## build a linar model for data that has outliers

```python
lb = preprocessing.LabelBinarizer()
data['result'] = lb.fit_transform(data['Opportunity Result'])
```

## encode the rest of the categorical variables

```python
data.head()
```

```python
data.groupby('Competitor Type').size()
```

```python
le = preprocessing.LabelEncoder()
```

```python
data['Supplies_Subgroup_Encoded'] = le.fit_transform(data['Supplies Subgroup'])
data['Supplies_Group_Encoded'] = le.fit_transform(data['Supplies Group'])
data['Region_Encoded'] = le.fit_transform(data['Region'])
data['Route_To_Market_Encoded'] = le.fit_transform(data['Route To Market'])
data['Competitor_Type_Encoded'] = le.fit_transform(data['Competitor Type'])
```

```python
final_data = data.drop(['Supplies Subgroup', 'Supplies Group', 'Region', 'Route To Market', 'Competitor Type', 'Opportunity Result'], axis=1)
```

```python
final_data.head()
```

## prepare the data

```python
yVar = final_data['result']
xVar = final_data.loc[:, final_data.columns != 'results']
```

```python
X_train, X_test, y_train, y_test = train_test_split(xVar, yVar, test_size=0.2)
print (X_train.shape, y_train.shape)
print (X_test.shape, y_test.shape)
```

## build the model without removing any outliers

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
```
